```python
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
```

```python
# load ground truth data
true_A = pd.read_csv("../data/TestData_A.csv", index_col=0)
true_A = true_A.iloc[:, 1:]  # ignore aggregate load column
true_A_values_scaled = true_A.values / true_A.max(axis=0).values
```

```python
# load your data

Groupletter = 'X'

test = True
if test:
    pred_A = pd.read_csv("../data/DisaggregatedLoad_Template.csv", index_col=0)
else:
    pred_A = pd.read_csv(f"Loads_Group_{Groupletter}.csv", index_col=0)

pred_A_values_scaled = pred_A.values / true_A.max(axis=0).values
```

```python
rmse_sum = np.sum(
    mean_squared_error(true_A_values_scaled, pred_A_values_scaled, multioutput="raw_values")
)
```
